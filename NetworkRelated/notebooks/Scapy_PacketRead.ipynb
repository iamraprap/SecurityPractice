```python
from scapy.all import IP, Raw, rdpcap

packets = rdpcap('pcap\Apple.Com.gz.pcap')
print(packets)
```

```python
checking_first_packet = True
client_ip_addr = None
server_ip_addr = None

ssl_packets = []
messages = []
```

```python
for packet in packets:
    if checking_first_packet:
        client_ip_addr = packet[IP].src
        checking_first_packet = False
        print("client_ip_addr:%s" % (client_ip_addr))
    else:
        if packet[IP].src != client_ip_addr:
            server_ip_addr = packet[IP].src
            print("server_ip_addr:%s" % (server_ip_addr))
    try:
        if (packet[Raw]):
            ssl_packets.append(packet)
            print("packet:%s" % (packet))
    except IndexError:
        pass
```
